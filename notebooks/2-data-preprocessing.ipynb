## Preprocessing of the "domy" dataset

```python
# to access sibling directories
import sys
import os
sys.path.append(os.path.abspath(".."))

from src.preprocessor import DataPreprocessor
import pandas as pd
import yaml

path_input = '../data/raw/domy.csv'
path_output = '../data/processed/domy.csv'
path_config = '../config/preprocessor.yaml'

df = pd.read_csv(path_input)
with open(path_config, 'r') as f:
    config = yaml.safe_load(f)

p = DataPreprocessor(df, config)
p.preprocess()
p.preprocessed_data.to_csv(path_output, index=False)
```
